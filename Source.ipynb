```python
from textblob.classifiers import NaiveBayesClassifier
for x in range(10):
    fp = 0
    fn = 0
    tp = 0
    tn = 0
    
    with open("data/train_test/train_"+str(x)+".csv",) as f:
        cl = NaiveBayesClassifier(f,format="csv",errors = "ignore")
    
    data_test = pd.read_csv("data/train_test/test_"+str(x)+".csv", names = ["User_Review","Polarity"])
    for index,row in data_test.iterrows():
        polarity = row['Polarity']
        text = row['User_Review'].decode('utf-8').strip()
        prediction = cl.classify(text)
    
    
    
    
        if(prediction == 'pos' and polarity == 'pos'):
            tp+=1
        
        elif(prediction == 'pos' and polarity == 'neg'):
            fp+=1
        
        elif(prediction == 'neg' and polarity == 'neg'):
            tn+=1
        
        elif(prediction == 'neg' and polarity == 'pos'):
            fn+=1
    
    
    accuracy = ((tp+tn)/(tp+tn+fp+fn))
    precision = (tp/(tp+fp))
    recall = (tp/(tp+fn))

    fm = ((precision*recall)/(precision+recall))
    fm = 2*fm
    print "Fold -",x
    print "Accuracy: ", accuracy
    print "Precision: ", precision
    print "Recall: ",recall
    print "F-Measure: ",fm
```

```python
for x in range(10):
    fp = 0
    fn = 0
    tp = 0
    tn = 0
    
    
    data_test = pd.read_csv("data/train_test/test_"+str(x)+".csv", names = ["User_Review","Polarity"])
    for index,row in data_test.iterrows():
        polarity = row['Polarity']
        text = row['User_Review'].decode('utf-8').strip()
        blob = TextBlob(text, analyzer=NaiveBayesAnalyzer())
        blob = blob.sentiment
        prediction = blob[0]
        
    
    
    
    
        if(prediction == 'pos' and polarity == 'pos'):
            tp+=1
        
        elif(prediction == 'pos' and polarity == 'neg'):
            fp+=1
        
        elif(prediction == 'neg' and polarity == 'neg'):
            tn+=1
        
        elif(prediction == 'neg' and polarity == 'pos'):
            fn+=1
    
    
    accuracy = ((tp+tn)/(tp+tn+fp+fn))
    precision = (tp/(tp+fp))
    recall = (tp/(tp+fn))

    fm = ((precision*recall)/(precision+recall))
    fm = 2*fm
    print "Fold -",x
    print "Accuracy: ", accuracy
    print "Precision: ", precision
    print "Recall: ",recall
    print "F-Measure: ",fm
```
